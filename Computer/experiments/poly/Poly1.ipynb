```python
class FixedLines(object):
    def __init__(self, left_poly, left_orient, right_poly, right_orient):
        self.left_poly = left_poly
        
        self.right_poly = right_poly
    def process(image):
        left_fit = {
            'poly': self.left_poly,
            'orient': self.left_orient,
            'xy': self._calculate_xy(self.left_poly, self.left_orient)
        }
        right_fit = {
            'poly': self.right_poly,
            'orient': self.right_orient,
            'xy': self._calculate_xy(self.right_poly, self.right_orient)
        }
        return ((left_fit, right_fit))
```

```python
local_paths = img_paths[10:20]
i = 1
total = len(local_paths)*4
cols = 4
fig = plt.figure(figsize=(15, (total/cols)*3))
for img_path in local_paths:
    print(os.path.basename(img_path))
    img_raw = cv2.imread(img_path)
    img = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
    warped = cv2.warpPerspective(img, M, (TARGET_WIDTH, TARGET_HEIGHT))

    # Display the original, undistorted, and warped image
    a = fig.add_subplot((total/cols)+1, cols, i)
    plt.imshow(img)
    i+=1

#     preprocessed = preprocess_line_finding(img_raw, mtx, dist, M, ((106, 11, 11), (255, 109, 209)))
    preprocessed = preprocess_line_finding(img_raw, M,
                                           sobel=True, inrange=((106, 14, 76), (255, 255, 255)))

    a = fig.add_subplot((total/cols)+1, cols, i)
    plt.imshow(preprocessed, cmap="gray")
    i+=1
    
    # Find lines
    a = fig.add_subplot((total/cols)+1, cols, i)
    finder = FindLinesSlidingWindows(nwindows=40, window_minpix=10, left_search_margin=10, right_search_margin=5,
                                     v_hist_crop_top=0.1, error_threshold=80, window_patience=24, window_empty_px=0,
                                     lr_start_slack=0.1, error_top_percentile=70, center_importance=2,
                                     lines=(1, 0, 1, 0), vert_x_adjust=(0, 0))
    fits = finder.process(preprocessed)
    final_img = annotate_with_lines(warped, fits, keep_background=True)
    plt.imshow(final_img)
    i += 1
    
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    a = fig.add_subplot((total/cols)+1, cols, i)
    final_img = cv2.warpPerspective(final_img, Minv, (final_img.shape[1], final_img.shape[0]))
    plt.imshow(final_img)
    i += 1
```
